## Kaj je program?
Program je zaporedje ukazov, ki računalniku povedo kaj mora narediti. Ukazi se izvajajo po vrsti, najprej prvi, nato drugi in tako naprej.

V starih časih so za to kar uporabljali tiskalnike, ki so rezultat natisnili na papir, dandanes pa imamo za to namenjene zaslone. Za izpis v programskem jeziku Python uporabljamo ukaz print.

```python
print("Pozdravljen, svet!")
```

## Spremenljivke
Spremenljivke si lahko predstavljamo kot nekakšne škatle z določenim imenom, katere imajo notri shranjeno določeno vrednost. Spremenljivko nastavimo z znakom `=`. Na levi strani enačaja se nahaja ime spremenljivke, na desni pa vrednost, na katero jo nastavljamo. Primer deklaracije spremenljivke:


```python
a = 3
```

**Pozor!** Imena spremenljivk morajo slediti določenim pravilom:
1. imena spremenljivk lahko vsebujejo vse črke **angleške** abecede, številke ter podčrtaje `_`,
2. ime spremenljivke ne sme vsebovati presledkov (primer: `moje ime` ni veljavno),
3. ime spremenljivke se ne sme začeti s številko (primer: `1sprem` ni veljavno),
4. imena spremenljivk so občutljiva na velikost črk (primer: `a` ni enaka spremenljivka kot `A`)

Vrednost spremenljivke lahko izpišemo s pomočjo ukaza print, samo znotraj oklepajev namesto besedila vstavimo ime spremenljivke.

```python
print(a)
```

**3. naloga** Ustvari 2 spremenljivki z vrednostima 3 in 7 ter ju izpiši.

Spremenljivke razdelimo v več kategorij, ki jih imenujemo tipi. Vsak tip ima določene svoje lastnosti, lahko pa tudi ustvarjamo lastne tipe spremenljivk, a o tem prihodnjič.

Nekaj tipov spremenljivk:
- integer (celo število)
- float (število s plavajočo vejico (realno))
- boolean (binarna spremenljivka z dvema vrednostima: `True` ali `False`)
- string (besedilo)

Tip spremenljivke se v Pythonu samodejno nastavi glede na vrednost na katero nastavimo.

```python
a = 3 # integer (celo število)
b = 3.141 # float (realno število)
c = True # boolean (true ali false)
d = "Neko besedilo" # string (besedilo)
```

**Pozor!** Za zapis decimalnih števil se uporablja **decimalna pika** in **ne vejica**.

### Izpis besedila in spremenljivk
Velikokrat želimo narediti lepši izpis za končnega uporabnika, zato bi radi k rezultatu dodali nekaj spremljevalnega besedila. Če uporabljamo Python 3, lahko to enostavno naredimo s tako imenovanim formatiranjem besedil. 

```python
print(f"Spremenljivka a ima vrednost {a}.")
```

Formatirano besedilo je označeno s črko `f` pred narekovaji. Na mesto, kamor želimo vstaviti spremenljivko vpišemo `{` in `}`, med njiju pa vstavimo ime spremenljivke.

### Operacije med spremenljivkami
Med spremenljivkami lahko opravljamo večino osnovnih matematičnih operacij:
- seštevanje `+`
- odštevanje `-`
- množenje `*`
- deljenje `/`
- ostanek pri deljenju `%`
- potenciranje `**`

Prednosti med operacijami so enake kot pri matematiki, lahko pa jih prilagodimo z uporabo oklepajev.

Primer uporabe operatorjev:

```python
a = 3 # ustvarimo in nastavimo a
b = 9 # ustvarimo in nastavimo b
c = a + b # ustvarimo c in ga nastavimo na vsoto a in b
print(c) # izpišemo c
```

### Vpis števil
Ukaz input vedno vrne spremenljivko tipa besedilo. Včasih želimo vpisano besedilo pretvoriti v številko, da lahko potem naprej računamo. To naredimo z ukazi za pretvarjanje med tipi spremenljivk.

```python
a = int(input("Vpiši število: ")) # Vpraša po vpisu a ter pretvori besedilo v celo število
b = float(input("Vpiši realno število: ")) # Vpraša po vpisu b ter pretvori besedilo v realno število

print(a * b) # Izpiše zmožen a in b
```

## Dodatne knjižnjice
Ker je Python odprtokoden velik del jezika dopolnjuje skupnost. Dodatne zbirke ukazov imenujemo knjižnjice. Z njimi si poenostavimo programiranje, saj so drugi (velikokrat pametnejši) programerji za nas že napisali algoritme. Tukaj si bomo ogledali knjižnjico `math`, v kateri dobimo dodatne matematične funkcije.

```python
import math # Dodajanje knjižnjice v program
```

Nekako po standardu se vsi import stavki vedno vpisujejo na vrh programa. 

Nekaj primerov ukazov iz knjižnjice `math`:

```python
print(math.pi) # Izpiše pi
print(math.sin(math.pi/2)) # izpiše sin(pi/2) - vrednost v sinusu je v radianih
print(math.sqrt(16)) # izpiše kvadratni koren od 16
```
